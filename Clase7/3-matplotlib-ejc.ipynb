```python
#USANDO NUMPY PARA GENERAR GRAFICAS A PARTIR DE UN ARRAY y funciones matematicas

import numpy as np


x = np.linspace(-2,2,500)
print(f"Array de x:\n{x}")
y = x**2

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y)
plt.title("square fuction") #titulo de la grafica
plt.xlabel('X')  #label for x axis
plt.ylabel('Y = X**2')  #nombre para el eje Y
plt.grid(True)   #mostrar rejilla de cuadricula
plt.show()      #muestra la gr√°fica en pantalla
```

```python
#SOBRESPONER GARFICAS Y CAMBIAR ESTILOS SW LINea

import numpy as np


x = np.linspace(-2,2,500)
print(f"Array de x:\n{x}")
y = x**2
Y2 = x +1  #Linea que se va a graf

plt.plot(x,y,'b--',x,Y2,'r.-')   #Grafica con dos curvas: una para y=x^2 y otra para y=x+1omillas es el color en ingles con la inicial
plt.show()
```

```python
#SEPARANDO DOS FUNCIONES CAD UNA EN UN PLOT

import numpy as np


x = np.linspace(-2,2,500)
# print(f"Array de x:\n{x}")
y = x**2    #Funci√≥n que se va a graficar en el eje y
Y2 = x +1  # funcion de la Linea que se va a graficar

plt.plot(x,y,'b--')   
plt.plot(x,Y2,'r.-')
plt.show()

```

```python
import numpy as np


x = np.linspace(-2,2,500)
# print(f"Array de x:\n{x}")
y = x**2    #Funci√≥n que se va a graficar en el eje y
y2 = x + 1  # funcion de la Linea que se va a graficar

plt.plot(x,y,'b--', label = "x**2")   
plt.plot(x,y2,'r.-', label = "x+1")
plt.legend(loc="best") #ubica la leyenda en la mejor localizacion de la grafica, como las convenciones d elos mapas
plt.show()
```

```python
#CREAR SUBPLOt PARA PONER UNA GRAFICA EN UN DETERMINADO ESPACIO DE UN GRID, ALGO ASI COMO POSICIONAR ELEMENTOS EN CSS GRID

import numpy as np


x = np.linspace(-2,2,500)
# print(f"Array de x:\n{x}")
y = x**2    #Funci√≥n que se va a graficar en el eje y
y2 = x + 1  # funcion de la linea que se va a graficar

plt.subplot(1,2,1)   #creamos una subfigura el primer valor es las filas rows, el segundo las columnas y el tercero le enterga la posiciona ocupar
plt.plot(x,y,'b--')   
plt.subplot(1,2,2)   #creamos una subfigura el primer valor es las filas rows, el segundo las columnas y el tercero le enterga la posiciona ocupar
plt.plot(x,y2,'r.-')
plt.show()
```

```python
#Scatter Plots sirve para graficar puntos de dispersion, otro tipo de grafica para estadistica 

#importacion de random para generar arrrays  aleatorios y matplotlib para hacer el gr√°fico.
import numpy as np
import matplotlib.pyplot as plt

#generamos un array con numeros aleatorios                                                  
from numpy.random import rand

x,y = rand(2,100)
plt.scatter(x,y)
plt.show
```

```python
#ESTABLECIENDO DOS SCATTER POR SEPARADOS CON COLORES DISTINTOS

from numpy.random import rand

x,y = rand(2,100)
x2,y2 = rand(2,100)
plt.scatter(x,y, c = 'red')
plt.scatter(x2, y2, c='blue')
plt.show
```

```python
#GENERANDO HISTOGRAMAS CON HIST

data = [1,1.1,1.8,2,2.1,3.2,3,3,3,3]

plt.subplot(211)
plt.hist(data, bins=10, alpha=0.7)
plt.show
plt.savefig('histograma.png')
```
